# 4x4x4 - Training Graphs

In this notebook we compare the training of PPO with and without heuristic action masking.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load data
mean_reward_default = pd.read_csv('ep-rew-mean-4x4x4-default.csv')
mean_reward_optimized = pd.read_csv('ep-rew-mean-4x4x4-optimized.csv')
mean_reward_heuristic = pd.read_csv('ep-rew-mean-4x4x4-heuristics.csv')
entropy_loss_default = pd.read_csv('entropy-loss-4x4x4-default.csv')
entropy_loss_optimized = pd.read_csv('entropy-loss-4x4x4-optimized.csv')
entropy_loss_heuristic = pd.read_csv('entropy-loss-4x4x4-heuristics.csv')
```

```python
# merge mean reward
mean_reward_default.index = mean_reward_default['Step']
mean_reward_optimized.index = mean_reward_optimized['Step']
mean_reward_heuristic.index = mean_reward_heuristic['Step']

# plot mean reward
plt.plot(mean_reward_default['Value'].rolling(5).mean(), label='Default')
plt.plot(mean_reward_heuristic['Value'].rolling(5).mean(), label='Heuristic')
plt.plot(mean_reward_optimized['Value'].rolling(5).mean(), label='Optimized')
plt.title('4x4x4 - Mean Reward over Time')
plt.xlabel('Step')
plt.ylabel('Mean Reward')
plt.grid()
plt.xlim(0, 3000000)
plt.legend(loc='lower right')
plt.ylim(min(mean_reward_default['Value'].rolling(5).mean().min(),
             mean_reward_heuristic['Value'].rolling(5).mean().min(),
             mean_reward_optimized['Value'].rolling(5).mean().min()), 14.7);
```

```python
# merge entropy loss
entropy_loss_default.index = entropy_loss_default['Step']
entropy_loss_optimized.index = entropy_loss_optimized['Step']
entropy_loss_heuristic.index = entropy_loss_heuristic['Step']

# plot entropy loss
plt.plot(entropy_loss_default['Value'].rolling(5).mean(), label='Default')
plt.plot(entropy_loss_heuristic['Value'].rolling(5).mean(), label='Heuristic')
plt.plot(entropy_loss_optimized['Value'].rolling(5).mean(), label='Optimized')
plt.title('4x4x4 - Entropy Loss over Time')
plt.xlabel('Step')
plt.ylabel('Entropy Loss')
plt.grid()
plt.xlim(0, 3000000)
plt.legend(loc='lower right')
plt.ylim(min(entropy_loss_default['Value'].rolling(5).mean().min(),
             entropy_loss_heuristic['Value'].rolling(5).mean().min(),
             entropy_loss_optimized['Value'].rolling(5).mean().min()), 0);
```
